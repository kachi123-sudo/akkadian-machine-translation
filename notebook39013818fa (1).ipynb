```python
import re

def normalize_akkadian(text):
    text = text.lower()
    text = re.sub(r"[^\w\s\-]", "", text)  # remove punctuation
    text = text.replace("š", "s")
    text = text.replace("ṭ", "t")
    text = text.replace("ā", "a")
    text = text.replace("ī", "i")
    return text
```

```python
def translate_word_by_word(text, dictionary):
    words = normalize_akkadian(text).split()
    translated = []

    for w in words:
        translated.append(dictionary.get(w, w))

    return " ".join(translated)
```

```python
def apply_templates(text):
    text = text.replace("thus says", "Thus says")
    text = text.replace("tablet", "tablet")
    text = text.replace("when", "When")
    return text
```

```python
def translate(text):
    out = translate_word_by_word(text, AKK_DICT)
    out = apply_templates(out)
    return out
```

```python
import pandas as pd
import torch

from datasets import Dataset
from transformers import (
    T5ForConditionalGeneration,
    AutoTokenizer,
    DataCollatorForSeq2Seq,
    Trainer,
    TrainingArguments
)
```

```python
train_df = pd.read_csv(
    "/kaggle/input/deep-past-initiative-machine-translation/train.csv"
)
test_df = pd.read_csv(
    "/kaggle/input/deep-past-initiative-machine-translation/test.csv"
)
```

```python
train_ds = Dataset.from_pandas(
    train_df[["transliteration", "translation"]]
)
```

```python
model_name = "google/byt5-small"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)
```

```python
remove_unused_columns=False
```

```python
def preprocess(batch):
    inputs = [
        "translate Akkadian to English: " + t
        for t in batch["transliteration"]
    ]

    model_inputs = tokenizer(
        inputs,
        truncation=True,
        padding="max_length",
        max_length=256,
        text_target=batch["translation"],
    )

    return model_inputs
```

```python
train_ds = train_ds.map(
    preprocess,
    batched=True,
    remove_columns=train_ds.column_names
)
```

```python
training_args = TrainingArguments(
    output_dir="./byt5-akkadian",
    per_device_train_batch_size=4,
    gradient_accumulation_steps=4,
    learning_rate=1e-4,
    num_train_epochs=6,
    fp16=True,
    logging_steps=100,
    save_steps=10_000,
    save_total_limit=1,
    report_to="none",
    remove_unused_columns=False   # ← REQUIRED
)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_ds,
    data_collator=DataCollatorForSeq2Seq(tokenizer, model),
)
```

```python
trainer.train()
```

```python
def generate_translation(text):
    input_ids = tokenizer(
        "translate Akkadian to English: " + text,
        return_tensors="pt",
        truncation=True,
        max_length=256
    ).input_ids.to(model.device)

    outputs = model.generate(
        input_ids,
        max_length=256,
        num_beams=5,
        early_stopping=True
    )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
translations = [
    generate_translation(t)
    for t in test_df["transliteration"]
]
```

```python
submission = pd.read_csv(
    "/kaggle/input/deep-past-initiative-machine-translation/sample_submission.csv"
)

submission["translation"] = translations
submission.to_csv("submission.csv", index=False)
```
